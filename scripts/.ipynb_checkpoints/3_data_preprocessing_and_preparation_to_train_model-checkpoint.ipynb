**METHOD**
- We can divide `IMAGE` into smaller `sub-regions` or `patches`. These patches can be retrieved or extracted from the existing image by applying `sliding` windows or by randomly selecting `regions` of the image.
- We will use `butteo` library to generate `PATCHES` from which we can `group` them into Batches. These `Batches` are then feed into our CNN model during training.
- One major advantage of `generating patches` allows us to reduce the computation memory that is required to process the image data

### Avoiding Data Leakage:
- We need to be careful while generating patches/sub-regions from the selected region, it is important to be aware since original image can be sampled multiple times, if overlaps are used. This means patches are not always independent of each other and might partially overlap. This is especially important to take into account when creating the test set, as it is essential that it is not contaminated with training data.
- We can easily create the patches by using the array_to_patches function from the buteo library while specifying the size of the resulting arrays.

### TRAIN TEST SPLIT
- we need to be careful with data leakage possibilities when splitting the data. Here, data leakage occurs when `future` data are fed into training the model
- A machine learning model needs to be trained on a set of data to learn from it. However, to choose the best model or tune hyperparameters, we need a way to evaluate the model's performance on data it has not seen before. By splitting data into separate train and test sets, we can select the best model and hyperparameters. We will set aside 10% of the data to use for testing.

- To make sure the test set is completely independent from the train set, `we will split the data before making the patches.`

- The best practice in machine learning tasks is to have a validation set as well. The model is not trained on the validation set, is used to evaluate the performance of a model during training. The validation set can be used to avoid overfitting, which happens when a model fits too closely to the training data and performs poorly on new data.

### STEP: splitting the datasets
- randomly selecting 10% of data to hold out and use for testing
- this step allows to process rather than on `patch` level. Because performin splits on `patch` level can contaminate as SAME pixels can appear in both train and test

- patch size: 32*32, we use this patch size to reduce the amount of RAM required to run the model

```python
COLORMODE = "#38383838" ## darkmode

# Custom function for creating subplots with a specified face color
def custom_subplots(*args, facecolor=COLORMODE, size=None, **kwargs):
    fig, axes = plt.subplots(*args, **kwargs)

    if size is not None:
        fig.set_size_inches(size[0], size[1])

    # Make sure axes is a list of axes objects
    if not isinstance(axes, Iterable) or isinstance(axes, np.ndarray) and axes.ndim == 0:
        axes = [axes]

    for ax in axes:
        ax.set_xticks([])
        ax.set_yticks([])

        for spine in ax.spines.values():
            spine.set_edgecolor('black')
            spine.set_linewidth(1)

    # Reduce the distance between the plotted images in the width
    plt.subplots_adjust(wspace=0.05)

    # Change the background color of the whole plot to grey
    fig.set_facecolor(facecolor)  # RGBA tuple for grey, 100% transparent if supported

    axes = axes[0] if len(axes) == 1 else axes

    return fig, axes
```

### Generate Image Patches by Image type and store as list
- using `buteo` library to convert patches from the `tiff` files
- First we convert the `tif` files to array, then we will create `patches` from the generate array file

```python
%%time
## read and order the TILES in temp. folder
for image in tqdm(zip(paths_labels, paths_s1, paths_s2, paths_dem), total= len(paths_labels),
                  ncols= 120):

    ## retrieve the elements zipped in `image`
    path_label, path_s1, path_s2, path_dem= image

    ## Get the name and number of PATCHES
    label_name= os.path.splitext(os.path.basename(path_label))[0]
    img_idx= int(label_name.split("_")[1])

    ## Get the data from tiles
    arr_label= beo.raster_to_array(path_label, filled= True, fill_value= 0.0)

    ## Handle any potential errors
    np.clip(arr_label, 0.0, 100.0, out= arr_label)
    arr_label[np.isnan(arr_label)] = 0.0

    ## Read the tiles
    arr_s1= beo.raster_to_array(path_s1, filled= True, fill_value=0.0)
    arr_s2= beo.raster_to_array(path_s2, filled= True, fill_value= 0.0)
    arr_dem= beo.raster_to_array(path_dem, filled= True, fill_value= 0.0)

    ## Generated patches
    patches_label = beo.array_to_patches(arr_label, PATCH_SIZE, n_offsets=N_OFFSETS)
    patches_s1 = beo.array_to_patches(arr_s1, PATCH_SIZE, n_offsets=N_OFFSETS)
    patches_s2 = beo.array_to_patches(arr_s2, PATCH_SIZE, n_offsets=N_OFFSETS)
    patches_dem = beo.array_to_patches(arr_dem, PATCH_SIZE, n_offsets=N_OFFSETS)

    # Sanity check to ensure that the right images were chosen.
    assert patches_label.shape[0:3] == patches_s1.shape[0:3] == patches_s2.shape[0:3] == patches_dem.shape[0:3], "Patches do not align."

    ## Based on the ID; seperate out the train and test list
    if img_idx in test_indices:
        testing_label.append(patches_label)
        testing_s1.append(patches_s1)
        testing_s2.append(patches_s2)
        testing_dem.append(patches_dem)
    else:
        training_label.append(patches_label)
        training_s1.append(patches_s1)
        training_s2.append(patches_s2)
        training_dem.append(patches_dem)
```

### Merge PATCHES together on axis->0

```python
# Merge the patches back together
training_label = np.concatenate(training_label, axis=0)
training_s1 = np.concatenate(training_s1, axis=0)
training_s2 = np.concatenate(training_s2, axis=0)
training_dem = np.concatenate(training_dem, axis=0)

testing_label = np.concatenate(testing_label, axis=0)
testing_s1 = np.concatenate(testing_s1, axis=0)
testing_s2 = np.concatenate(testing_s2, axis=0)
testing_dem = np.concatenate(testing_dem, axis=0)

print("Training Labels: ", training_label.shape, training_s1.shape, training_s2.shape, training_dem.shape)
print("Test Labels:     ", testing_label.shape, testing_s1.shape, testing_s2.shape, testing_dem.shape)
```

## Normalization on Input Pixels
- Normalization is a common preporcessing step in computer vision tasks
- Refers to process of `rescaling` the pixel values of an image to have a `common` scale
- `REASON`: to reduce the impact of the scale of input features on the learning process of the model

1. **Sentinel-1**: initially expressed in `dB`. The data is first clipped to the range(-5 dB, 35dB). Normalizing the value between the range(0, 1) where we subtract the mean and divide by the standard deviation.
<br><br>
2. **Sentinel-2**: These images are unit16 with values ranging from (0 to 65535) which are called digital numbers (DN). The physical value derived from this is the reflectance defined as DN/10.000. The refelectance is usually between 0-1 but can be higher due to surface or cloud effects.  By convention Sentinel 2 data is clipped to the range (0,10.000). Afterwards it is normalised to (0,1) as well.
<br><br>
3. **CopDEM:** The channels in the DEM are already normalised. The elevation (4th channel) is normalised with respect to the heighest point on earth which is  mt. Everest. The slope (3rd band) is also normalised with respect to a slope of 90°. The first two bands contain the orientation of the slope. Because the orientation is a cyclical variable (0° = 360°), we need two channels to encode and  normalise this information. This can be done by storing the sine of the orientation in the first band and the cosine in the second band. To keep the range between (0,1) the final result is:

$$ \text{channel 1} = [sin(\text{orientation})+1]/2 $$
$$ \text{channel 2} = [cos(\text{orientation})+1]/2 $$


## STEP to normalize each IMAGE types;
- **Sentinel-1** requires extra step, since it ranges from ~(-35 to 5)
- **Sentinel-2** data is normalized by dividing with 10000.0
- **DEM** has been already normalized

```python
# Scalers also return a dictionary of the stats, so the same scaler can be reused.
s1_train, _statdict = beo.scaler_truncate(training_s1, -35.0, 5.0)
s1_test, _statdict = beo.scaler_truncate(testing_s1, -35.0, 5.0)

s2_train, _statdict = beo.scaler_truncate(training_s2, 0.0, 10000.0)
s2_test, _statdict = beo.scaler_truncate(testing_s2, 0.0, 10000.0)

# If the labels are not already float32, cast them to float32.
labels_train = training_label.astype(np.float32, copy=False)
labels_test = testing_label.astype(np.float32, copy=False)
```

## MODEL building
1. Start by building simple convolutional network that can achieve significant results
2. Model takes input of `image` containing 9 `spectral bands` from Sentinel-2 (which has been normalized by multiplying with 10000.00) and split up into patches of 32*32
3. Output: this is built of density of `pixels` ranging between 0-100.   regression task as we are trying to predict specific values.

### Model architecture:
- The architecture is simple.
  - First there is an **encoder** block where the number of channels is increased to 128. Because of the padding, the resolution of the output channels is unchanged.
  - In the **decoder** block, there are extra convolutions that bring the number of channels `back` to a single one that contains the building density.

**NOTE:** To help the model we force the values to lie between in the range 0-100 since any other values will never be correct.

### Create class object for nn model
- Here, in this tutorial, we create a simple model, there are `NO` **bottlenecks, skip connections, attention layers or other modern features.**
- Further we perform hypertuning like changing the network structure which can provide more good results
- Adding `RESIDUAL BLOCK` with **InceptionResNet and ConvNext** for modern convolutional architecutres

```python
class SimpleConvNet(nn.Module):
    def __init__(self, in_channels, output_min, output_max):
        super(SimpleConvNet, self).__init__()
        self.output_min= output_min
        self.output_max= output_max

      ## ENCODER; this will increase the number of channels to 128, padding will keep the 
      ## resoultion same
        self.encoder= nn.Sequential(
            nn.Conv2d(in_channels, 64, kernel_size= 3, padding=1),
            nn.ReLU(inplace= True),
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 128, kernel_size= 3, padding= 1),
            nn.ReLU(inplace= True),
            nn.BatchNorm2d(128)
        )

     ## DECODER; with extra convolutions it will bring the number of CHANNELS
     ## back to SINGLE one which contains the building density
        self.decoder= nn.Sequential(
            nn.Conv2d(128, 64, kernel_size= 3, padding= 1),
            nn.ReLU(inplace= True),
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 1, kernel_size= 3, padding= 1)
        )

    ## econde then decode the input
    def forward(self, x):
        x= self.encoder(x)
        x= self.decoder(x)

        x= torch.clamp(x, self.output_min, self.output_max)
        return x
```

### Define the model
- Channel bands for Sentinel-2 includes:9

```python
## define the model; for Sentinel-2
input_channels= 9

## Labels will be in range(0, 100), we can set the network
model = SimpleConvNet(input_channels, 0.0, 100.0)

## Constants for the model. We can adjust the parameters in future
EPOCHS= 10
BATCH_SIZE= 16
LEARNING_RATE= 0.001
```

### Conventions: Based on different data format

### Using Torch to create `dataset` object
- Here, `DataLoader` class works by creating an ITERABLE dataset object and iteration occurs it in `batches`, which are then fed into MODEL for processing
- This can be generated via numpy, PyTorch tensors, etc.

```python
class NumpyDataset(Dataset):
    def __init__(self, x_train, y_train, data_is_channel_last= False):
        if data_is_channel_last:
            x_train= beo.channel_last_to_first(x_train)
            y_train= beo.channel_last_to_first(y_train)

        self.x_train= torch.from_numpy(x_train).float()
        self.y_train= torch.from_numpy(y_train).float()

    def __len__(self):
        return len(self.x_train)

    def __getitem__(self, index):
        x= self.x_train[index]
        y= self.y_train[index]

        return x, y
```

## Metrics:
- To measure and evaluate the performance of a model on a given task. Metrics are used to quantify how **well** a model is performing and to compare the performance of different models

### Regression
- We are trying to predict the **density of buildings** at a given `pixel`. A very common metric for regression taks is the Mean Squarred Error (MSE). For a given image the MSE is calcualte as follows:

$$MSE = \frac{1}{n} \sum_{i=1}^n (y_{pred,i} - y_{label,i})^2 $$

Where $n$ is the number of pixels in the image, $y_{pred,i}$ is the i-th pixel from the **prediction** and  $y_{label,i}$ is the i-th pixel from the **label**. <br>

In a similar fashion there is the Mean Average Error (MAE) and Root Mean Squared Error (RMSE) which are expressed in the same units as the label:
$$MAE = \frac{1}{n} \sum_{i=1}^n |y_{pred,i} - y_{label,i}| $$
<br>
$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^n (y_{pred,i} - y_{label,i})^2}.$$


## TRAIN the model
- While training the model our `loss function` optimization will be MSE. We can further inspect the training loss during the training our model
- We can move our model to GPU if available for much faster computation

### Functionalities of Torch in brief:
- `Forward function:` calcuation process as the output layers values from the input data
- `MSELoss()`; we will try to optimize the loss function; with the outputs and targets
- `device`, often model requiers powerful computation to handle large data, so this will move the model computation to GPU usage if available
- `backward loss`; this process is counting the changes in the `weights` by using the gradient decent algorithm
- `optimizer.zero_grad()`; this is used to clear the gradients of all parameters in a model

```python
%%time
import torch.optim as optim

## our loss function: MSE
criterion= nn.MSELoss()

## optimizer for our loss function
optimizer= optim.Adam(model.parameters(), lr= LEARNING_RATE)

## move model to GPU if available
device= torch.device("cuda:0" if torch.cuda.is_available() else "cpu" )
model.to(device)

## Training Loop
for epoch in range(EPOCHS):
    
    running_loss= 0.0

    ## Initialize the steps during training
    train_bar= tqdm(dataloader, total= len(dataloader), ncols= 120)

    for i, (inputs, targets) in enumerate(train_bar):
        ## move inputs and targets to DEVICE(GPU)
        inputs, targets= inputs.to(device), targets.to(device)

        ## zero gradients
        optimizer.zero_grad()

        ## forward pass
        outputs= model(inputs)

        ## compute the loss
        loss= criterion(outputs, targets)

        ## backward pass
        loss.backward()

        ## update the weights
        optimizer.step()

        ## show the stats
        current_loss= loss.item()

        running_loss += current_loss
        mean_loss= running_loss / (i+1)

        ## update the progress of training bar
        train_bar.set_description(f"Epoch: {epoch + 1:03d} / {EPOCHS:03d}")
        print_dict= {"loss": f"{mean_loss:4f}"}

        train_bar.set_postfix(print_dict)
```

## Brief information on the model architecture before Loading the Model
- Define the model architecture as used during training.

### Some of the specific **Reasons:**
- `compatibility:` model architecture needs to match the one used during training. Loading a saved model into a diff. architecutre can lead **to errors.**, as `SAVED weights` and `layers` might not align with expected structure<br><br>
- `Parameter Initialization:` When defining a new model with same architecutre, it typically initalizes the model's parameters (weights and biases) with default values. . If you load a saved model after defining a new architecture, the loaded weights and biases will overwrite the initialized ones. This can result in the loss of valuable pre-trained information, and the model may not perform as expected.<br><br>
- `Evaluation Mode` Some layers in deep learning models, such as Dropout and Batch Normalization, behave differently during training and inference. By defining the model architecture before loading, you can explicitly set the model to evaluation mode (using model.eval()) or training mode (using model.train()) as needed. This ensures that these layers behave correctly during inference.<br><br>
- `Optimizer Compatibility` If you're also loading the optimizer state along with the model, it's important that the optimizer is associated with the same model architecture. Otherwise, you may encounter issues when trying to update the model's parameters during further training.

### Function to predict LARGE image
- using library `buteo` to predict array
- Automatically allows to predict LARGE images and write off lenghty code

```python
def predict(arr):
    swap = beo.channel_last_to_first(arr)
    as_torch = torch.from_numpy(swap).float()
    on_device = as_torch.to(device)
    predicted = model(on_device)
    on_cpu = predicted.cpu()
    as_numpy = on_cpu.numpy()
    swap_back = beo.channel_first_to_last(as_numpy)

    return swap_back
```

### Trimming the outliers
- We clip top and bottom 2% of the image array by `trimming` the data to normalize the outliers

```python
def clip_quantiles(arr, q_min= 0.02, q_max= 0.98):
    return np.clip( arr, 
                  np.nanquantile(arr, q_min), 
                  np.nanquantile(arr, q_max))
```

### Transormation of Sentinel-2 to RGB image
- Transformming the raster image of 16 bits to 8 bits
- To transform 16 bits to 8 bit RGB rendering we normalize the LAYER by dividing with MAX value from the image and round up to nearest integer

```python
def render_s2_to_rgb(arr):
    ## if nodata values, we CAST all values to zeros
    if np.ma.isMaskedArray(arr):
        arr= np.ma.getdata(arr.filled(0))

    ## Select only Red, Blue, Green from the BANDS
    rgb_slice= arr[:, :, 0:3]

    ## Clip the data by quantiles, so the RGB render is not Stretched to outliers
    ## this can result in dark images
    for c in [ 0, 1 , 2]:
        rgb_slice[:, :, c]= clip_quantiles(rgb_slice[:, :, c])

    ## Currently, the slice has 16 bits, we want to transform to 8 bit RGB render
    ## 1. we normalize the LAYER by dividing with MAX value from the image
    ## 2. we multiply by 255 (max of 8 bit) which ranges from 0-255, this will normalize
    ## within the RGB range
    for c in [0, 1, 2]:
        rgb_slice[:, :, c]= (rgb_slice[:, :, c] / rgb_slice[:, :, c].max()) * 255.0

    ## Third, we will ROUND-UP to nearest integer and cast to uint8
    rgb_slice= np.rint(rgb_slice).astype(np.uint8)

    return rgb_slice
```
